```python
import pandas as pd
import numpy as np
```

```python
fifaM = pd.read_csv('/content/drive/MyDrive/Data Science archivos/fifaM.csv')
fifaM
```

```python
fifaM['X'].astype(str)
```

```python
fifaM_ = fifaM[fifaM['X'].astype(str).apply(lambda x: '€' not in x)]
fifaM_
```

```python
mi_diccionario = {ord('€'): None }

fifaM_['Release.Clause'] = fifaM_['Release.Clause'].str.translate(mi_diccionario)
fifaM_
```

```python
fifaM_['Release.Clause'] = fifaM_['Release.Clause'].str.replace('K', '000')
```

```python
mi_diccionario_2 = {ord('€'): None ,ord('.'): None, ord('K'): 000, ord('M'): '000000'}
fifaM_['Release.Clause'] = fifaM_['Release.Clause'].str.translate(mi_diccionario_2)  
fifaM_
```

```python
fifaM_ = fifaM_[fifaM_['Release.Clause'].notna()]
fifaM_
```

```python
mean = np.mean(fifaM_['Release.Clause'].astype(int))
```

En este apartado se obtiene la media de la columna 'Release.Clause'

Obtener la media de 'Release.Clause'

```python
print(f'La media es de: {mean}')
```
